# Linear Regression Model

Linear Regression is a regression model that finds a linear relationship between independent and dependent variables. It assumes that the relation ship is linear and the residuals are normally distributed.

* Scaling data is a common practice in machine learning to ensure that all features contribute equally to the model.

```python
# Scaling the dataset to fit the model
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

* Dividing test-train data

* Scratch Implementation of Linear Regression

```python
# y = mx + c
class LinearRegression:
    def __init__(self, learning_rate=0.001, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None # m
        self.bias = None # c

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y)) # partial derivative
            db = (1 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.lr * dw # update weight
            self.bias -= self.lr * db # update bias

    # y_pred = X. weight + bias
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

* Firstly, using consturctor, learning rate of gradient descent, no. of iterations for gd, weight and bias(mean and intercept) of linear regression model are initialized.
* Fit method trains the model to input featires X and target values y. No. of samples and features determines X.shape.Initializes each weights as zero and self bias to zero.
* Loop runs on n_iter no. of times for gradient descent optimization to reduce loss function and update model parameters. y_pred are predicted target values.
dw are gradient loss function respect to weight and db are gradient loss function with respect to bias.
* Then, predict method predicts target values from trained model taking input features X.

```python
# Train the model
regressor = LinearRegression(learning_rate=0.01, n_iters=1000)
regressor.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = regressor.predict(X_test)
```

```python
# Calculate R^2 (coefficient of determination)
ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)
ss_res = np.sum((y_test - y_pred) ** 2)
r2 = 1 - (ss_res / ss_tot)
print("R^2:", r2)
```

```python
# Calculate RMSE
rmse = np.sqrt(np.sum((y_test - y_pred) ** 2) / len(y_test))
print("RMSE:", rmse)
```

#Ridge Regression


Another model as ridge regression model is taken into reference for evaluating the model.
Ridge regression extends linear regression by tacking multicollinearity(situation where 2 or more independent features are highly correlated) problem by introducing a regularization parameter.

L2 regularization to prevent overfitting.
alpha is the regularization parameter that controls the strength of regularization. It is used on partial derivative of geadient descent with respect to weights.

```python
# y = (alpha)mx + c
class RidgeRegression:
    def __init__(self, learning_rate=0.01, n_iters=1000, alpha=1.0):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.alpha = alpha  # Regularization parameter
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias
            dw = (2 / n_samples) * (np.dot(X.T, (y_pred - y)) + self.alpha * self.weights)
            db = (2 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    # y_pred = X. weight + bias
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# Train model
ridge_model = RidgeRegression(learning_rate=0.01, n_iters=1000, alpha=1.0)
ridge_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_ridge = ridge_model.predict(X_test)

# Evaluate the model
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
r2_ridge = r2_score(y_test, y_pred_ridge)

# Print evaluations
print("Ridge Regression Model:")
print(f"Root Mean Squared Error: {rmse_ridge}")
print(f"R-squared Score: {r2_ridge}")
```

*(With similar scoring in linear and ridge regression, the does not seem to be case of multicollinearity in the dataset)*

# Feature Selection

Wrapper method is used as a feature selection techique.

Here, feature selection introduces a better analysis of data by selecting the best feature within the dataset that corresponds to better performance of the model.

Within Wrapper method, I selected Sequential Forward Selection method so that it can iteratively select significant one and only include the necessities.

```python
selected_features = set()

best_scores = []

num_features_to_select = 3

while len(selected_features) < num_features_to_select:
    best_feature = None
    best_score = 0

    for feature_idx in range(X_train.shape[1]):
        if feature_idx not in selected_features:
            current_features = list(selected_features) + [feature_idx]

            reg_model = LinearRegression()
            reg_model.fit(X_train[:, current_features], y_train) #direct indexing

            y_pred = reg_model.predict(X_test[:, current_features])

            score = mean_squared_error(y_test, y_pred)

            if score > best_score:
                best_score = score
                best_feature = feature_idx

    selected_features.add(best_feature)

    best_scores.append(np.sqrt(best_score))

print("Selected Features:", selected_features)
print("Best Scores at Each Step:", best_scores)
```

Firstly, an empty set of selected features is selected alongside a list of best scores. Then the count of features to select is declared. An iteration is initialized such that it iterates till the count of feature selection criteria isnt complete. Within loop, best features and best scores are initialized to 0 then the feature iteration begins. It loops around all features and checks if feature is already selected or not. It trains regression model based on selected features and evalutaes the performance with mse on test set. If current feature improves or has the best score then it updates `best score` and `best features` variables.

# Implementing the CV and Feature selection methods

* For linear regression model

> * Pipeline declaration for a smoother seqential execution of feature selection and regression. Selects feature as SFS with (num_features_to_select) based on the performance of the LinearRegression model.
* Hyper Parameters are defined.
* Initialized gridSelectionCV with defined pipeline, params_grid and 5-fold cross validation.
* Fit gridCV for training model and set best_params and train model again with best parameters.
* Selected features are then extracted.
* Calculate mse and r^2 to evaluate best model and print the details.


```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('feature_selection', SequentialFeatureSelector(LinearRegression(), n_features_to_select=num_features_to_select, direction='forward')),
    ('regression', LinearRegression())
])

params_grid = {
    'features_to_select': [0,1, 5]  # Selected from above calculations
}

grid_search = GridSearchCV(pipeline, params_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

pipeline.set_params(**best_params)
pipeline.fit(X_train, y_train)

selected_features = pipeline.named_steps['feature_selection'].get_support(indices=True)

y_pred = pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Best Parameters:", best_params)
print("Selected Features:", selected_features)
print("MSE:", mse)
print("R^2 Score:", r2)
```

* For ridge regression model

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SequentialFeatureSelector

pipeline = Pipeline([
    ('feature_selection', SequentialFeatureSelector(Ridge(), n_features_to_select=num_features_to_select, direction='forward')),
    ('regression', Ridge())
])

params_grid = {
    'features_to_select': [0,1,5]  # Number of features to select
}

grid_search = GridSearchCV(pipeline, params_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

pipeline.set_params(**best_params)
pipeline.fit(X_train, y_train)

# Get selected features
selected_features = pipeline.named_steps['feature_selection'].get_support(indices=True)

# Get predictions
y_pred = pipeline.predict(X_test)

# Calculate MSE and RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Best Parameters:", best_params)
print("Selected Features:", selected_features)
print("Root Mean Squared Error:", rmse)
```
